### Функция для удаления слов содержайщих цифры

```python
def remove_words_with_digit(text):
    return re.sub(r"\b\S*\d\S*\b", '', text)
```

```python
df['toxic'] = df['toxic'].replace(1.0, 'neg')
df['toxic'] = df['toxic'].replace(0.0, 'pos')
```

## CountVectorizer

```python
cv = CountVectorizer(stop_words=stop_russian_words)
```

```python
X_train_countvec = cv.fit_transform(X_train)
```

```python
pd.DataFrame(pd.Series(cv.get_feature_names())).head()
```

```python
X_test_countvec = cv.transform(X_test)
```

```python
countvector_word_df = pd.DataFrame(pd.Series(cv.get_feature_names()), columns = {'word'})
countvector_word_df
```

### Функция для сравнения результатата модели

```python
def report(model):
    preds = model.predict(X_test_countvec)
    print(classification_report(y_test, preds))
    conf_mat = confusion_matrix(y_test, preds, labels= model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat,display_labels= model.classes_)
    disp.plot()
```
